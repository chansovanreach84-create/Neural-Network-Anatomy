# Build a Standard MLP
* Task: Create a class inheriting from torch.nn.Module. Build a Multilayer Perceptron (MLP) with one input layer, two hidden layers, and an * output layer. Write the forward method to pass data through it. Pass a dummy tensor through to ensure the output shape matches your expectations.

Why: Familiarizes you with the standard API for building network architectures.


```python
import torch
import torch.nn as nn
class StandardMLP(nn.Module):
    def __init__(self, input_size, hidden1_size, hidden2_size, output_size):
        super(StandardMLP, self).__init__()
        
        # Define the layers
        self.fc1 = nn.Linear(input_size, hidden1_size)   # Input to Hidden 1
        self.relu1 = nn.ReLU()
        
        self.fc2 = nn.Linear(hidden1_size, hidden2_size) # Hidden 1 to Hidden 2
        self.relu2 = nn.ReLU()
        
        self.fc3 = nn.Linear(hidden2_size, output_size)  # Hidden 2 to Output
        
    def forward(self, x):
        # Pass data sequentially through the layers
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.fc3(out)
        return out

INPUT_SIZE = 10
HIDDEN1 = 32
HIDDEN2 = 16
OUTPUT_SIZE = 2

model = StandardMLP(INPUT_SIZE, HIDDEN1, HIDDEN2, OUTPUT_SIZE)
print(model)

# Create dummy data: batch_size = 5, features = 10
dummy_input = torch.randn(5, INPUT_SIZE)

# Forward pass
dummy_output = model(dummy_input)

# Verify the shapes
print(f"Input shape:  {dummy_input.shape}")   # Expected: torch.Size([5, 10])
print(f"Output shape: {dummy_output.shape}")  # Expected: torch.Size([5, 2])
```
